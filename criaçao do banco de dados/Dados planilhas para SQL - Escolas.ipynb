```python
import pandas as pd
```

```python
# Nomes planilhas
planilha_inse_2011_a_2013 = 'Indicador_INSE_por_Escola_2011_a_2013.xlsx'
planilha_inse_2015 = 'INSE_2015.xlsx'
planilha_inse_2019 = 'INSE_2019_ESCOLAS.xlsx'
planilha_inse_2021 = 'INSE_2021_escolas.xlsx'
```

```python
df_inse_2011_a_2013 = pd.read_excel(planilha_inse_2011_a_2013)
df_inse_2015 = pd.read_excel(planilha_inse_2015)
df_inse_2019 = pd.read_excel(planilha_inse_2019)
df_inse_2021 = pd.read_excel(planilha_inse_2021)
```

```python
df_uniques_subset_2011_a_2013 = df_inse_2011_a_2013.drop_duplicates(subset=['COD_ESCOLA', 'COD_MUNICIPIO', 'ID_REDE', 'ID_LOCALIZACAO', 'NOME_ESCOLA'])
df_uniques_subset_2011_a_2013 = df_uniques_subset_2011_a_2013.rename(columns={'COD_ESCOLA': 'CO_ESCOLA', 'COD_MUNICIPIO': 'CO_MUNICIPIO', 'ID_REDE': 'TP_DEPENDENCIA', 'ID_LOCALIZACAO': 'TP_LOCALIZACAO'})
#df_uniques_subset_2011_a_2013.set_index('CO_ESCOLA', inplace=True)
```

```python
df_uniques_subset_2015 = df_inse_2015.drop_duplicates(subset=['CO_ESCOLA', 'CO_MUNICIPIO', 'TP_DEPENDENCIA', 'TP_LOCALIZACAO', 'NOME_ESCOLA'])
#df_uniques_subset_2015.set_index('CO_ESCOLA', inplace=True)
```

```python
df_uniques_subset_2019 = df_inse_2019.drop_duplicates(subset=['CO_ESCOLA', 'CO_MUNICIPIO', 'TP_DEPENDENCIA', 'TP_LOCALIZACAO', 'NOME_ESCOLA'])
#df_uniques_subset_2019.set_index('CO_ESCOLA', inplace=True)
```

```python
df_uniques_subset_2021 = df_inse_2021.drop_duplicates(subset=['ID_ESCOLA', 'CO_MUNICIPIO', 'TP_TIPO_REDE', 'TP_LOCALIZACAO', 'NO_ESCOLA'])
df_uniques_subset_2021 = df_uniques_subset_2021.rename(columns={'ID_ESCOLA': 'CO_ESCOLA', 'TP_TIPO_REDE': 'TP_DEPENDENCIA', 'NO_ESCOLA': 'NOME_ESCOLA'})
#df_uniques_subset_2021.set_index('CO_ESCOLA', inplace=True)
```

```python
# CONCATENANDO DATAFRAMES
columns_to_select = ['CO_ESCOLA', 'CO_MUNICIPIO', 'TP_DEPENDENCIA', 'TP_LOCALIZACAO', 'NOME_ESCOLA']
concatenated_df = pd.concat([df_uniques_subset_2011_a_2013[columns_to_select], df_uniques_subset_2015[columns_to_select], df_uniques_subset_2019[columns_to_select], df_uniques_subset_2021[columns_to_select]]).drop_duplicates(subset='CO_ESCOLA', keep='last')
```

### Obtendo Latitude e Longitude da Escola

```python
file_escolas_centro_oeste = '../DadosEscolas/Escolas_regiao_centro_oeste.csv'
file_escolas_nordeste = '../DadosEscolas/Escolas_regiao_nordeste.csv'
file_escolas_norte = '../DadosEscolas/Escolas_regiao_norte.csv'
file_escolas_sudeste = '../DadosEscolas/Escolas_regiao_sudeste.csv'
file_escolas_sul = '../DadosEscolas/Escolas_regiao_sul.csv'
```

```python
df_centro_oeste = pd.read_csv(file_escolas_centro_oeste, delimiter=';')
df_nordeste = pd.read_csv(file_escolas_nordeste, delimiter=';')
df_norte = pd.read_csv(file_escolas_norte, delimiter=';')
df_sudeste = pd.read_csv(file_escolas_sudeste, delimiter=';')
df_sul = pd.read_csv(file_escolas_sul, delimiter=';')
```

```python
columns_to_select = ['Código INEP', 'Latitude', 'Longitude']
escolas_detalhes_df = pd.concat([df_centro_oeste[columns_to_select], df_nordeste[columns_to_select], df_norte[columns_to_select], df_sudeste[columns_to_select], df_sul[columns_to_select]]).drop_duplicates()
escolas_detalhes_df.set_index('Código INEP', inplace=True)
```

```python
escolas_detalhes_df
```

#### Montando script

```python
import psycopg2
```

```python
# Connect to the PostgreSQL database
connection = psycopg2.connect(
    host="localhost",
    database="postgres",
    user="postgres",
    password="senha"
)

cursor = connection.cursor()
```

```python
# MONTANDO SCRIPT

values = ''
for indice, linha in concatenated_df.iterrows():
    
    id = linha['CO_ESCOLA']
    id_municipio = linha['CO_MUNICIPIO']
    id_dependencia = linha['TP_DEPENDENCIA']
    id_localizacao = linha['TP_LOCALIZACAO']
    nome_escola = linha['NOME_ESCOLA']   
    
    if id in escolas_detalhes_df.index and not pd.isna(escolas_detalhes_df.loc[id]['Latitude']) and not pd.isna(escolas_detalhes_df.loc[id]['Longitude']) :
        latitude = escolas_detalhes_df.loc[id]['Latitude']
        longitude = escolas_detalhes_df.loc[id]['Longitude']

        inicio_script_escolas = 'INSERT INTO saeb.tb_escola(id_escola, fk_id_municipio, fk_id_dependencia_administrativa, fk_id_localizacao, no_escola, nu_latitude, nu_longitude) VALUES '
        sql_insert = inicio_script_escolas + f"({id}, {id_municipio}, {id_dependencia}, {id_localizacao}, '{nome_escola}', {latitude}, {longitude});"

    else:
        inicio_script_escolas = 'INSERT INTO saeb.tb_escola(id_escola, fk_id_municipio, fk_id_dependencia_administrativa, fk_id_localizacao, no_escola) VALUES '
        sql_insert = inicio_script_escolas + f"({id}, {id_municipio}, {id_dependencia}, {id_localizacao}, '{nome_escola}');"

    cursor.execute(sql_insert)

    # Commit the transaction
    connection.commit()

# Close the cursor and connection
cursor.close()
connection.close()
```
