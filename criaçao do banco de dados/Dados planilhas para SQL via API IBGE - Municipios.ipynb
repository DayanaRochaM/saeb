```python
import requests
import pandas as pd
```

```python
# Nomes planilhas 
url_municipios = 'https://servicodados.ibge.gov.br/api/v1/localidades/municipios?orderBy=nome'
```

## Obtendo dados para insert de Municípios

```python
# Fazendo a requisição GET
response_municipios = requests.get(url_municipios)
if response_municipios.status_code == 200:
    # Imprimindo o conteúdo da resposta
    municipios = response_municipios.json()
else:
    # Imprimindo o código de status em caso de falha
    print(f"A requisição falhou com o código de status: {response_municipios.status_code}")
```

#### Buscando informações se o Município é de área urbana ou rural

```python
# Nomes planilhas
planilha_inse_2011_a_2013 = 'Indicador_INSE_por_Escola_2011_a_2013.xlsx'
planilha_inse_2015 = 'INSE_2015.xlsx'
planilha_inse_2019 = 'INSE_2019_ESCOLAS.xlsx'
planilha_inse_2021 = 'INSE_2021_escolas.xlsx'
```

```python
df_inse_2011_a_2013 = pd.read_excel(planilha_inse_2011_a_2013)
df_inse_2015 = pd.read_excel(planilha_inse_2015)
df_inse_2019 = pd.read_excel(planilha_inse_2019)
df_inse_2021 = pd.read_excel(planilha_inse_2021)
```

```python
df_uniques_subset_2011_a_2013 = df_inse_2011_a_2013.drop_duplicates(subset=['COD_MUNICIPIO', 'ID_AREA'])
df_uniques_subset_2011_a_2013.set_index('COD_MUNICIPIO', inplace=True)
```

```python
df_uniques_subset_2015 = df_inse_2015.drop_duplicates(subset=['CO_MUNICIPIO', 'ID_AREA'])
df_uniques_subset_2015.set_index('CO_MUNICIPIO', inplace=True)
```

```python
df_uniques_subset_2019 = df_inse_2019.drop_duplicates(subset=['CO_MUNICIPIO', 'ID_AREA'])
df_uniques_subset_2019.set_index('CO_MUNICIPIO', inplace=True)
```

```python
df_uniques_subset_2021 = df_inse_2021.drop_duplicates(subset=['CO_MUNICIPIO', 'TP_CAPITAL'])
df_uniques_subset_2021.set_index('CO_MUNICIPIO', inplace=True)
```

```python
# IGUALANDO OS NOMES DAS COLUNAS PARA PODER CONCATERNAR OS DATAFRAMES
df_uniques_subset_2011_a_2013 = df_uniques_subset_2011_a_2013.rename(columns={'COD_MUNICIPIO': 'CO_MUNICIPIO'})
df_uniques_subset_2021 = df_uniques_subset_2021.rename(columns={'TP_CAPITAL': 'ID_AREA'})
```

```python
# CONCATENANDO DATAFRAMES
concatenated_df = pd.concat([df_uniques_subset_2011_a_2013, df_uniques_subset_2015, df_uniques_subset_2019, df_uniques_subset_2021]).drop_duplicates()
```

```python
# ADICIONANDO ID_AREA AOS MUNICIPIOS
for municipio in municipios:
    if municipio['id'] in concatenated_df.index:
        municipio.update({'id_area':concatenated_df.loc[municipio['id']]['ID_AREA'].max()})
```

#### Montando script

```python
# MONTANDO SCRIPT
inicio_script_municipios = 'INSERT INTO saeb.tb_municipio(id_municipio, fk_id_unidade_federativa, fk_id_area, no_municipio) VALUES '

values = ''
for municipio in municipios:
    
    id = municipio['id']
    # Uma micro-região geralmente é composta por municípios que pertencem ao mesmo estado, por isso usaremos ela para obter a chave UF primaria
    id_UF = municipio['microrregiao']['mesorregiao']['UF']['id']
    id_area = municipio['id_area']
    nome_municipio = municipio['nome'].replace("'", "''") if "'" in municipio['nome'] else municipio['nome']
    values += f"({id}, {id_UF}, {id_area},'{nome_municipio}'),"
values = values[:-1] #Retirando vírgula da última linha

# Montando script completo e definindo arquivo de saída
sql_script = inicio_script_municipios + values + ';'
caminho_arquivo_sql = 'SCRIPTS/script_insert_municipios.sql'

# Escrevendo os dados no arquivo
with open(caminho_arquivo_sql, 'w') as arquivo:
    arquivo.write(sql_script)
```
