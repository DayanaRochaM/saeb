#### Lendos todas as planilhas

```python
import pandas as pd
```

```python
# Nomes planilhas
planilha_inse_2011_a_2013 = 'Indicador_INSE_por_Escola_2011_a_2013.xlsx'
planilha_inse_2015 = 'INSE_2015.xlsx'
planilha_inse_2019 = 'INSE_2019_ESCOLAS.xlsx'
planilha_inse_2021 = 'INSE_2021_escolas.xlsx'
```

```python
df_inse_2011_a_2013 = pd.read_excel(planilha_inse_2011_a_2013)
df_inse_2015 = pd.read_excel(planilha_inse_2015)
df_inse_2019 = pd.read_excel(planilha_inse_2019)
df_inse_2021 = pd.read_excel(planilha_inse_2021)
```

```python
import psycopg2

# Connect to the PostgreSQL database
connection = psycopg2.connect(
    host="localhost",
    database="postgres",
    user="postgres",
    password="senha"
)

cursor = connection.cursor()
```

### 2011 e 2013

```python
columns_to_select = ['COD_ESCOLA', 'QTD_ALUNOS_INSE', 'INSE - VALOR ABSOLUTO']
df_uniques_subset_2011_a_2013 = df_inse_2011_a_2013[columns_to_select].drop_duplicates(subset=columns_to_select)
df_uniques_subset_2011_a_2013.set_index('COD_ESCOLA', inplace=True)
```

```python
df_uniques_subset_2011_a_2013
```

```python
indice_inse_escola = 1

for indice, linha in df_uniques_subset_2011_a_2013.iterrows():

     #INSERT tb_inse_classificacao#
    # DECIDIR QUAL A CLASSIFICACAO
    valor = linha['INSE - VALOR ABSOLUTO']
    id_classificacao = -1 #apenas um valor inicial
    if(0<= valor <=30):
        id_classificacao = 1
    elif(30< valor <=40):
        id_classificacao = 2
    elif(40< valor <=50):
        id_classificacao = 3
    elif(50< valor <=60):
        id_classificacao = 4
    elif(60< valor <=70):
        id_classificacao = 5
    elif(70< valor <=80):
        id_classificacao = 6
    elif(80< valor):
        id_classificacao = 21
    else:
        print(f"Erro: sem classificacao para a escola {indice}")
    
    #INSERT tb_inse_escola
    inicio_script = 'INSERT INTO saeb.tb_inse_escola(nu_ano_inicio, nu_ano_fim, qtd_alunos, fk_id_escola, nu_valor_inse, fk_id_classificacao) VALUES '
    sql_insert = inicio_script + f"(2011, 2013, {linha['QTD_ALUNOS_INSE']}, {indice}, {linha['INSE - VALOR ABSOLUTO']}, {id_classificacao});"
    
    #faz insert aqui
    cursor.execute(sql_insert)
    connection.commit()

    indice_inse_escola += 1
```

```python
indice_inse_escola
```
